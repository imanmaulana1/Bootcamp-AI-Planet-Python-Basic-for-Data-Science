## Task 2
Write code to print characters from a string that are present at an even index number. For example, if your input string is "python", the output characters will be (p,t,o).

```python
string = input()
for i in range(0,len(string)):
  if i%2==0 :
    print(string[i])
  
```

## Task 3

Write a function that calculates the average of the elements of the list: 

elements = [5, 7, 4, 9, 2]


```python
list = [5,7,4,9,2]
hasil = 0
for num in list:
  hasil+=num
  
avg = hasil/len(list)
print(avg)
```

## Task 4
Write a code that takes a year as input and determines whether it is a leap year or not.

```python
yr=int(input('Year: '))
if yr%100==0: 
  if yr%400==0:
    print(yr,'is leap year')
  else:
    print (yr,'is not leap year')
else:
  if yr%4==0:
    print (yr,'is leap year')
  else:
    print (yr,'is not leap year')
```

## Task 5
Given a list of numbers (integers), find the sum of second maximum and second minimum in the following list.

l = [21, 6, 12, 57, 35, 61]

```python
l = [21, 6, 12, 57, 35, 61]

#copy list from var l and then assign to var l2
l2 = l.copy()

#sort by asc from list l
l.sort()

#sort by desc from list l2
l2.sort(reverse=True)

#get index 1 from lists
second_maximum = l[1]
second_minimum = l2[1]

result = second_maximum + second_minimum

print(l)
print('Second Maximum:',second_maximum)
print('Second Minimum:',second_minimum)
print('\nAnswer',second_maximum,'+',second_minimum,'=',result)
```
